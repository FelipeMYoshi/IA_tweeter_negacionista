```python
from sklearn.cluster import KMeans
import pandas as pa
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from google.colab import drive
from sklearn import svm, datasets
import seaborn as sb
import matplotlib.pyplot as pp
import numpy as np
import unicodedata
```

```python
datatwitters = pa.read_csv('/content/drive/My Drive/Projeto Integrador 3B/DataSet_PI3B.csv', sep=';',encoding='utf-8', names=['Texto'])
twitters = datatwitters['Texto'].tolist()
for n in range(len(twitters)):
  twitters[n] = twitters[n].upper()
```

```python
datapalavras = pa.read_csv('/content/drive/My Drive/Projeto Integrador 3B/palavras_PI3B.csv', sep=';',encoding='utf-8', names=['Palavras'])
palavras = datapalavras['Palavras'].tolist()
for n in range(len(palavras)):
  palavras[n] = palavras[n].upper()
```

```python
dataset = pa.DataFrame()

for n in range(len(palavras)):
  vetor = []
  for j in range(len(twitters)):
    if palavras[n] in twitters[j].split():
      valor = 1
    if palavras[n] not in twitters[j].split():
      valor = 0
    vetor.append(valor)
  dataset.insert(loc=0, column='nome', value=vetor, allow_duplicates=True)

dataset.columns = palavras[::-1]

dataset
```

```python
dataset = dataset.rename(columns={'TRONBOSE': 'TROMBOSE'})
```

```python
dataset.to_csv("/content/drive/My Drive/Projeto Integrador 3B/dataset_binario.csv", index=False)
```

```python
from sklearn.model_selection import train_test_split

X = dataset
Y = pa.read_csv('/content/drive/My Drive/Projeto Integrador 3B/Alvo_PI3B.csv', sep=';',encoding='utf-8', names=['Alvo'])

X_treino, X_teste = train_test_split(X, test_size = 0.30, random_state = 15 )
Y_treino, Y_teste = train_test_split(Y, test_size = 0.30, random_state = 15)
```

```python
from sklearn.neural_network import MLPClassifier

classificador = MLPClassifier(hidden_layer_sizes=(500, 500), max_iter=10000000, activation='relu', 
                              solver='adam', learning_rate_init=0.01, random_state=1) 
classificador.fit(X_treino, Y_treino)
Y_predicao = classificador.predict(X_teste)
```

```python
from sklearn.neural_network import MLPClassifier

classificador = MLPClassifier(hidden_layer_sizes=(500, 500),
                              max_iter=10000000,
                              alpha = 1e-5, 
                              activation='logistic',
                              solver='lbfgs', 
                              random_state = 10, 
                              learning_rate = 'adaptive') 

classificador.fit(X_treino, Y_treino)
Y_predicao = classificador.predict(X_teste)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score
print('='*80)
matriz_confusao = confusion_matrix(Y_predicao, Y_teste)
cmtx = pa.DataFrame(matriz_confusao, index=(['IH: Negacionista', 'IH: Não Negacionista']),columns=(['IC: Negacionista', 'IC: Não Negacionista']))
print(cmtx)
acc = accuracy_score(Y_teste,Y_predicao)
print('='*80)
print('     Acurácia no Teste: {0:0.2f}%'.format(acc*100))
precisao = precision_score(Y_teste, Y_predicao, average='macro')
print('     Precisão do Teste: {0:0.2f}%'.format(precisao*100))
print('='*80)
```
